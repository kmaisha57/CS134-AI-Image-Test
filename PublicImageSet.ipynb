Add Gaussian and Pixelation Blur

```python
import os
import cv2
import numpy as np

# Paths
INPUT_FOLDER = "/content/drive/MyDrive/CS134/ffhq_subset"
BLURRED_FOLDER = "/content/drive/MyDrive/CS134/blurred_ffhq_strong"
os.makedirs(BLURRED_FOLDER, exist_ok=True)

def apply_multistep_blur(image):
    """
    Applies a combination of Gaussian blur, downsampling+upsampling,
    and mild JPEG-like degradation to simulate strong real-world blur.
    """
    # Step 1: Apply heavy Gaussian blur
    blurred = cv2.GaussianBlur(image, (75, 75), sigmaX=30)

    # Step 2: Downsample + Upsample (adds pixelation-like degradation)
    h, w = blurred.shape[:2]
    downscale_factor = 0.25  # Reduce size by 75%
    small = cv2.resize(blurred, (int(w * downscale_factor), int(h * downscale_factor)), interpolation=cv2.INTER_LINEAR)
    upsampled = cv2.resize(small, (w, h), interpolation=cv2.INTER_LINEAR)

    # Step 3: Mild JPEG-style degradation (optional)
    encode_param = [int(cv2.IMWRITE_JPEG_QUALITY), 40]  # Lower quality = stronger artifacts
    _, encoded_img = cv2.imencode('.jpg', upsampled, encode_param)
    degraded = cv2.imdecode(encoded_img, cv2.IMREAD_COLOR)

    return degraded

# Loop through files and apply blur
for filename in os.listdir(INPUT_FOLDER):
    if filename.lower().endswith((".png", ".jpg")):
        img_path = os.path.join(INPUT_FOLDER, filename)
        img = cv2.imread(img_path)

        if img is not None:
            final_blur = apply_multistep_blur(img)
            output_path = os.path.join(
                BLURRED_FOLDER,
                filename.replace(".png", "_blurred.png").replace(".jpg", "_blurred.jpg")
            )
            cv2.imwrite(output_path, final_blur)

print("Multi-step blur applied to all images.")
```

```python
import os
import cv2
import numpy as np
from gfpgan import GFPGANer
import torch
from google.colab import drive

# Mount your Drive
drive.mount('/content/drive2')

# Set paths
INPUT_FOLDER = "/content/drive/MyDrive/CS134/blurred_ffhq_strong"
OUTPUT_FOLDER = "/content/drive/MyDrive/CS134/restored_faces_ffhq"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# Load GFP-GAN
model_path = "/content/drive2/MyDrive/CS134/GFPGANv1.3.pth"
restorer = GFPGANer(
    model_path=model_path,
    upscale=1,
    arch="clean",
    channel_multiplier=2,
    bg_upsampler=None
)

# Process all blurred images
for filename in os.listdir(INPUT_FOLDER):
    if filename.endswith(".png") or filename.endswith(".jpg"):
        img_path = os.path.join(INPUT_FOLDER, filename)
        output_path = os.path.join(OUTPUT_FOLDER, f"restored_{filename}")
        image_bgr = cv2.imread(img_path)

        if image_bgr is None:
            print(f"Could not read: {img_path}")
            continue

        print(f"Restoring: {filename}")

        try:
            # Convert BGR (OpenCV) â†’ RGB (for model)
            image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

            # Apply GFP-GAN
            _, restored_img, _ = restorer.enhance(
                image_rgb,
                has_aligned=False,
                only_center_face=True,
                paste_back=True
            )

            if restored_img is None:
                print(f"No face detected in {filename}")
                continue

            if isinstance(restored_img, list):
                restored_img = restored_img[0]

            if isinstance(restored_img, torch.Tensor):
                restored_img = restored_img.cpu().numpy()

            if restored_img.max() <= 1.0:
                restored_img = np.clip(restored_img * 255.0, 0, 255)

            restored_img = restored_img.astype(np.uint8)

            # Convert RGB â†’ BGR for saving
            if restored_img.shape[-1] == 3:
                restored_img = cv2.cvtColor(restored_img, cv2.COLOR_RGB2BGR)

            # Save output
            cv2.imwrite(output_path, restored_img)
            print(f"Saved: {output_path}")

        except Exception as e:
            print(f"Error on {filename}: {str(e)}")

print("All faces processed!")

```

```python
import os
import cv2
import numpy as np
import torch
import lpips
from skimage.metrics import structural_similarity as ssim
from glob import glob

# Initialize LPIPS
loss_fn = lpips.LPIPS(net='alex')

# Paths
ORIGINAL_FOLDER = "/content/drive/MyDrive/CS134/ffhq_subset/"
RESTORED_FOLDER = "/content/drive/MyDrive/CS134/restored_faces_ffhq"

# Evaluation functions
def compute_psnr(img1, img2):
    mse = np.mean((img1 - img2) ** 2)
    return 20 * np.log10(255.0 / np.sqrt(mse)) if mse > 0 else float('inf')

# Normalize for LPIPS
def preprocess_for_lpips(img):
    img = cv2.resize(img, (64, 64))  # Resize for consistency
    img = img.astype(np.float32) / 255.0
    img = (img * 2) - 1  # Normalize to [-1, 1]
    return torch.tensor(img).permute(2, 0, 1).unsqueeze(0)

# Process images
results = []
restored_images = glob(f"{RESTORED_FOLDER}/*.jpg") + glob(f"{RESTORED_FOLDER}/*.png")
print(f"Found {len(restored_images)} images in {RESTORED_FOLDER}")

for restored_path in restored_images:
    base_name = os.path.basename(restored_path).replace("restored_", "").replace("_blurred", "")
    original_path = os.path.join(ORIGINAL_FOLDER, base_name)

    print(f"Comparing restored: {restored_path} original: {original_path}")

    if os.path.exists(original_path):
        img1 = cv2.imread(original_path)
        img2 = cv2.imread(restored_path)

        if img1 is None or img2 is None:
            print(f"Could not read image(s): {original_path} or {restored_path}")
            continue

        # Grayscale for SSIM
        img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
        img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
        if img1_gray.shape != img2_gray.shape:
            img2_gray = cv2.resize(img2_gray, (img1_gray.shape[1], img1_gray.shape[0]))

        ssim_score = ssim(img1_gray, img2_gray)

        # LPIPS
        img1_lpips = preprocess_for_lpips(img1)
        img2_lpips = preprocess_for_lpips(img2)
        lpips_score = loss_fn(img1_lpips, img2_lpips).item()

        # PSNR
        if img1.shape != img2.shape:
            img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))
        psnr_score = compute_psnr(img1, img2)

        results.append((base_name, ssim_score, lpips_score, psnr_score))

print("Evaluation complete!")
```
